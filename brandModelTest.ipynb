```python
def replace_func(match):
        # Recuperar o acrônimo correspondente em minúsculas
        acr = match.group().lower()
        # Retornar a definição completa do acrônimo
        return dict_brando.get(acr, acr)  # Retorna o acrônimo se não encontrado no dicionário


def remove_x000d(text):
    return text.replace('_x000d_', '')


def expand_acronyms(text):
# Compilar o padrão de regex com todos os acrônimos
    text = remove_x000d(text)
    pattern = re.compile(r'\b(' + '|'.join(dict_brando.keys()) + r')\b', re.IGNORECASE)

    # Substituir acrônimos no texto
    return pattern.sub(replace_func, text)


def transform_text(text):
    # Clean text
    text = text.replace(r'[^a-zA-Z0-9\s]', '',)
    text = text.lower()
    return text


def tokenize(text):
    # Tokenize text
    tokens = nltk.tokenize.word_tokenize(text)
    return tokens   


def remove_stopwords(tokens):
    # Remove stopwords from tokens
    stopwords = set(nltk.corpus.stopwords.words('portuguese'))
    filtered_tokens = [token for token in tokens if token not in stopwords]
    return filtered_tokens

def lemmatize(tokens):
    # Lemmatize tokens
    lemmatizer = nltk.stem.WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]
    return lemmatized_tokens


def preprocess_text(text):
    text = transform_text(text)
    text = expand_acronyms(text)
    tokens = tokenize(text)
    tokens = remove_stopwords(tokens)
    tokens = lemmatize(tokens)
    return tokens
```

```python
# Função para fazer previsões
def make_predictions(model, texts):
    predictions = []
    for text in texts:
        output = model(
            f"<s>[INST] {text} [/INST]",  # Formatação do prompt
            max_tokens=512,  # Gerar até 512 tokens
            stop=["</s>"],  # Token de parada
            echo=True  # Se deve ecoar o prompt
        )
        predictions.append(output)
    return predictions

# Fazer previsões
predictions = make_predictions(llm, texts_to_predict)
```
